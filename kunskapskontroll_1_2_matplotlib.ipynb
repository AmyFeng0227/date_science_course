### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
x = np.linspace(-4, 4, 100)
y = x**2

# the object-oriented approach
fig, ax = plt.subplots()
print(ax.plot(x, y))

```

```python
#pyplot/matlab approach
plt.plot(x, y)
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
"a figure is like a blank canvas where you can add plots."
"axes is the actual plot where you can draw points, lines"
"axis is a numerical scale along which data points are plotted. Each axis has a label, ticks, and tick labels to provide context about the data being displayed."
"basically everything visible on a figure is an Artist."
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4)
y = x**2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.title("a")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.plot(x, y)
```
